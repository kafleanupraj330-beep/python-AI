```python
#1- Draw the lines 
import numpy as np
import matplotlib.pyplot as plt 
x= np.linspace(-10,10,100)
y1= 2 * x+1
y2= 2 * x+2
y3 = 2 * x+3
plt.plot(x,y1, 'r-+', label='y= 2x+1')
plt.plot(x,y2, 'g-,', label='y= 2x+2')
plt.plot(x,y3, 'b-.', label='y= 2x+3')
plt.title('lines in the same figure.')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
Convert lengths from inches to centimeters and weights from pounds to kilograms.
Calculate the means of the lengths and weights.
Finally draw a histogram of the lengths.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#reaing csv file by panda
df = pd.read_csv("weight-height.csv")
#taking data from csv file from two colums
length = df["Height"].to_numpy()
weight= df["Weight"].to_numpy()
#conversion
length_cm = length*2.54
weight_kg = weight*0.45
#calculation 
mean_length_cm = np.mean(length_cm)
mean_weight_kg = np.mean(weight_kg)
print(f"mean of the length is:{mean_length_cm}cm")
print(f"mean of the weight is:{mean_weight_kg}kg")
#plotting diagram 
plt.hist(length, bins=30,color="b", edgecolor="k")
plt.title("Histogram")
plt.xlabel("Height(inches)")
plt.ylabel("y")
plt.show()

```

```python
#inverse matrix
# Matrix A
A = np.array([[1, 2, 3],
              [0, 1, 4],
              [5, 6, 0]])

#inverse of A
A_inv = np.linalg.inv(A)

print("Inverse of A:")
print(A_inv)

print("\nA × A⁻¹:")
print(A @ A_inv)

print("\nA⁻¹ × A:")
print(A_inv @ A)
```

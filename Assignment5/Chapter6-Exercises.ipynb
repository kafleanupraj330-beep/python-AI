```python
#Exercise 6.1
#Consider the dataset from data_banknote_authentication.csv.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
#1) Read data into a pandas dataframe
df = pd.read_csv('data_banknote_authentication.csv')
```

```python
#2) Pick the target variable y as weight in kilograms, and the feature variable X as height in centimeters.
x= df.drop(columns=['class'])
y= df['class']
```

```python
#3) Split the data into training and testing sets with 80/20 ratio.
x_train, x_test,y_train, y_test= train_test_split(x,y, test_size=0.20, random_state=20)
```

```python
#4) Use support vector classifier with linear kernel to fit to the training data.
linear_svc = SVC(kernel='linear')
linear_svc.fit(x_train, y_train)
```

```python
#5) Predict on the testing data and compute the confusion matrix and classification report.
y_pred_linear = linear_svc.predict(x_test)

print("Support Vector Classifier with Linear Kernel")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_linear))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_linear))
```

```python
#6) Repeat steps 3 and 4 for the radial basis function kernel.
rbf_svc = SVC(kernel='rbf')
rbf_svc.fit(x_train, y_train)
y_pred_rbf = rbf_svc.predict(x_test)

print("\nSupport Vector Classifier with RBF Kernel")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rbf))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_rbf))
```

```python
#7) Compare the two SVM models in your own words
Linear SVM: Through its miss-classification of just two examples, the linear model is able to demonstrate that the divisions among genuine 
and fake banknotes are virtually linear and separable through the original feature space.
RBF SVM: The ability of the RBF kernel to use a non-linear radial basis function enables it to project the data in a higher dimensional space,
which is what helps it classify everything with no mistakes.
```
